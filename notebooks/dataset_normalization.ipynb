```python
df = df.drop(df.columns[:2], axis=1)
```

```python
df.drop(df.columns[:4], axis=1, inplace=True)
```

```python
df.info()
```

```python
df.head(3)
```

```python
columns_to_concatenate = ["gender", "age", "positionName", 
                          "skills_cv", "driveLicenses", "experience", "country", "educationList",
                          "hardSkills_cv", "softSkills_cv", "workExperienceList",
                          "languageKnowledge_cv", "salary_cv"]
```

```python
for column in columns_to_concatenate:
    df[column] = df[column].fillna("No information")
```

```python
df["Full_resume"] = df[columns_to_concatenate].agg(lambda row: ", ".join([f"{col}: {row[col]}" for col in columns_to_concatenate]), axis=1)
```

```python
df.head(1)
```

```python
df = df[[df.columns[-1]]]
```

```python
df.head(1)
```

```python
df1 = df.iloc[:25000, :]
df2 = df.iloc[25000:50000, :]
```

```python
df1.to_csv("../datasets/train_normalized_1.csv", index=False, sep="|")
df2.to_csv("../datasets/train_normalized_2.csv", index=False, sep="|")
```
